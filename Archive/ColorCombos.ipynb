```python
## RANDOM SELECTOR
## USED FOR THE COMPLEXITY GRAPH INITIALLY WITH 3 CHOICES -- MODIFY AS NEEDED
##
import random
import apdocc
from apdocc import Colors, Combinations

# --- pick a random color combo safely ---
if isinstance(Combinations, dict):
    # randomly choose one of the defined color combinations
    combo_name = random.choice(list(Combinations.keys()))
    combo = Combinations[combo_name]
elif hasattr(Combinations, "random"):
    combo = Combinations.random()
else:
    # fallback to 3 random Colors if Combinations unavailable
    combo = [random.choice(list(Colors.values())) for _ in range(3)]

# Normalize to list of hex values
if isinstance(combo, dict):
    color_list = list(combo.values())
elif hasattr(combo, "colors"):
    color_list = [getattr(c, "hex", str(c)) for c in combo.colors]
elif isinstance(combo, (list, tuple)):
    color_list = [getattr(c, "hex", str(c)) for c in combo]
else:
    color_list = [getattr(combo, "hex", str(combo))]

# Ensure exactly 3 colors (cycle or trim)
levels = ["Low Complexity", "Medium Complexity", "High Complexity"]
while len(color_list) < len(levels):
    color_list += color_list
color_list = color_list[:len(levels)]

palette = {lvl: color_list[i] for i, lvl in enumerate(levels)}
```
